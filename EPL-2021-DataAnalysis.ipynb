```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read the dataset(csv)
df_epl = pd.read_csv('england-premier-league-players-2018-to-2019-stats.csv')
```

```python
df_epl
```

```python
# Check for null values
df_epl.isna().sum()
```

```python
# Create additional columns "MinPerMatch" and "goalPerMatch"
df_epl['MinPerMatch'] = (df_epl['minutes_played_overall'] / df_epl['appearances_overall'])
df_epl['goalPerMatch'] = (df_epl['goals_overall'] / df_epl['appearances_overall']).astype(float)
```

```python
# print first 5 records or pass an interger parameter df.head(n) to print the first n numbers
df_epl.head()
```

```python
#Total goals
total_goals = df_epl['goals_overall'].sum()
print(total_goals)
```

```python
#Total Penalty Goals
total_penalty_goals = df_epl['penalty_goals'].sum()
print(total_penalty_goals)
```

```python
#Total Penalty Missed
total_penalty_miss = df_epl['penalty_misses'].sum()
print(total_penalty_miss)
```

```python
#Penalty Attempts
penalty_attempts = total_penalty_goals + total_penalty_miss
print(penalty_attempts)
```

```python
#Pie chart for Penalty scored vs missed
plt.figure(figsize=(13,6))
data = [total_penalty_miss, total_penalty_goals]
labels = ['Missed', 'Scored']
color = sns.color_palette('Set1')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
#Total number of Forward players. df_epl[condition]  where condition: df_epl['position'] == 'Forward'

df_epl[df_epl['position'] == 'Forward'].head()
```

```python
# Number of nations with players in EPL
#option a
a = df_epl['nationality'].unique().size
#option b
b = np.size(df_epl['nationality'].unique())
print(a)
print(b)
```

```python
#Visual: Top 10 countries with players in EPL
TopNations = df_epl.groupby('nationality').size().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10,6), color=sns.color_palette('magma'))
```

```python
# Clubs with lagest number players (Top 5)
df_epl['Current Club'].value_counts().nlargest(5).plot(kind='bar', figsize=(8,6), color=sns.color_palette('Set2'))
```

```python
# Clubs with least number of players (Bottom 5)
df_epl['Current Club'].value_counts().nsmallest(5).plot(kind='bar', figsize=(8,6), color=sns.color_palette('Set1'))
```

```python
#Players based on age group
under25 = df_epl[(df_epl['age'] > 20) & (df_epl['age'] <= 25)]
under35 = df_epl[(df_epl['age'] > 25) & (df_epl['age'] <= 35) ]
above35 = df_epl[df_epl['age'] > 35]
data = np.array([under25['age'].count(), under35['age'].count(), above35['age'].count()])
mylabels = ['20 - 25yrs','25 - 35yrs','Above 35yrs']
plt.title('Age Grouping', fontsize=20)
plt.pie(data, labels=mylabels, autopct='%1f%%')
plt.show()
```

```python
# Particular age group in all the clubs
under35['Current Club'].value_counts().plot(kind='bar', figsize=(13,6), color=sns.color_palette('Set2'))
```

```python
# Players in particular age group in a single club
under25[under25['Current Club']=='Liverpool']
```

```python
# Average Age of Players in each club
plt.figure(figsize=(12,6))
sns.boxplot(x='Current Club', y='age', data = df_epl)
plt.xticks(rotation =90)
```

```python
#Total assist from each club
assist_by_clubs = pd.DataFrame(df_epl.groupby('Current Club', as_index=False)['assists_overall'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Current Club', y='assists_overall', data=assist_by_clubs.sort_values(by='assists_overall'))
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=75)
#plt.rcParams['figure.figsize'] = (10,6) # You can specify size if you want
plt.title('Clubs and their Total Assists', fontsize=20)
```
